### If you want to sort by descending order then you also have to use the DESC(ASC) keyword.


Syntax: SELECT columns FROM table ORDER BY column DESC;

### GET TOP ASOS with maximum current price to least

### Select only 10 items with very high price loss or discounted price

### Data Visualization for 10 highest discounts on ASOS DESIGN Brand

```python
AHD=pd.read_sql('''
                               SELECT *, ((LOSS/Previous_Price)*100) AS Discount_per FROM Asos_loss 
                               WHERE Name LIKE 'ASOS%' ORDER BY Discount_per DESC LIMIT 10
                               ''', asos20)
AHD
```

```python
df = (AHD.groupby('Discount_per')['Name'].value_counts().unstack()
df.plot.bar()
```
