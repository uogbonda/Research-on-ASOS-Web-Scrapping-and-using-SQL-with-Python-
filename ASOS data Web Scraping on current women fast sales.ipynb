### Put everything together. Do a loop through the results and append the data inside a list

```python
# Creating a loop

name=[]
brand=[]
current_price=[]
previous_price=[]
for results in result_items:
    # name
    name.append(results['name'])
    # brand
    brand.append(results['brandName'])
    # current price
    current_price.append(results['price']['current']['text'])
    # previous price
    previous_price.append(results['price']['previous']['text'])
```

```python
asos_data=pd.DataFrame({'Name':name,'Brand':brand,'Current Price':current_price,'Previous Price':previous_price})
```
